```python
import pandas as pd
import numpy as np
```

```python
    #series from list 

#string 
country = ['India','pakistan','america','iran','indonesia']
pd.Series(country)
```

```python
# Integer

runs = [13,24,53,45,21]
pd.Series(runs,dtype='int16')
```

```python
#coustom index 

marks = [45,34,54,23]
subject = ["maths","science","English","hindi"]

pd.Series(marks,index=subject)
```

```python
#setting name

marks = [45,34,54,23]
subject = ["maths","science","English","hindi"]

pd.Series(marks,index=subject,name="Vedanti ke marks")
```

```python
#series from dictionary 

marks = {'maths':89,
         'science':78,
         'hindi':56,
         'English':79}

marks_series = pd.Series(marks,name="Yash ke marks")
marks_series
```

```python
fruits = {"Apples":30,
         "bananas":21}

pd.Series(fruits)
print(fruits)
```
