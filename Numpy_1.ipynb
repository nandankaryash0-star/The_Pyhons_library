```python
#Stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)

#horizontal stack inc coloum
np.hstack((a4,a5,a2))
```

```python
#vertical stack inc rows 

np.vstack((a4,a5))
```
