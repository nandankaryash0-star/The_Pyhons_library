syntax wise datetime is very convenient
But the performance takes a hit while working with huge data. List vs Numpy Array
The weaknesses of Python's datetime format inspired the NumPy team to add a set of native time series data type to NumPy.
The datetime64 dtype encodes dates as 64-bit integers, and thus allows arrays of dates to be represented very compactly.

```python
#generate daily dates in a given range 

pd.date_range(start='2026/1/1',end='2026/3/31',freq='3D')
```

```python
#alternate days 
pd.date_range(start='2026/1/1',end='2026/3/31',freq='3D')
```

```python
#B -> business days
pd.date_range(start='2026/1/1',end='2026/3/31',freq='B')
```

```python
#W->one week per day
pd.date_range(start='2026/1/1',end='2026/3/31',freq='W-THU')
```

```python
#h -> Hourly timestamp
pd.date_range(start='2026/1/1',end='2026/3/31',freq='h')
```

```python
#MS -> Month Start 
pd.date_range(start='2026/1/1',end='2026/3/31',freq='MS')
```

```python
#ME -> Month End
pd.date_range(start='2026/1/1',end='2026/3/31',freq='ME')
```

```python
#YE -> year end 
pd.date_range(start='2026/1/1',end='2026/3/31',freq='YE')
```

```python
#using periods (number of results )
pd.date_range(start='2026/1/1',periods=25,freq='h')
```

```python
#to_datetime

#simple series example 

s = pd.Series(['2026/1/1','2026/1/2','2026/1/3'])
#pd.to_datetime(s).dt.year
#pd.to_datetime(s).dt.day_name()
pd.to_datetime(s).dt.month_name()
```

```python
#with errors 
s = pd.Series(['2026/1/1','2026/1/2','2026/161/432'])
pd.to_datetime(s,errors='coerce').dt.year
```

```python
df = pd.read_csv('expense_data.csv')
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['Date'].dt.year
df['Date'].dt.day_name()
df['Date'].dt.month_name()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Date'],df['INR'])
```

```python
df['day_name'] = df['Date'].dt.day_name()
```

```python
df.groupby('day_name')['INR'].mean().plot(kind='bar')
```

```python
df['day_month'] = df['Date'].dt.month_name()
df.groupby('day_month')['INR'].mean().plot(kind='bar')
```
