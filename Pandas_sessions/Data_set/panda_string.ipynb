```python
df = pd.read_csv('titanic.csv')
df['Name']
```

```python
#common functions 
#lower/upper/capitalized/title
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()

#len
df['Name'][df['Name'].str.len() == 82].values[0]

#strip
df['Name'].str.strip()
```

```python
#split -> get

df['lastname'] = df['Name'].str.split(',').str.get(0)
df.head()
```

```python
#last name 
df['Name'].str.split(',').str.get(0).str.split(' ')
```

```python
df[['Title','Firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1, expand=True)
df.head()
```

```python
df['Title'].value_counts()
```

```python
#replace

df['Title'].str.replace('Ms.','Miss.')
df['Title'].str.replace('Mlle.','Miss.')
```

```python
df['Title'].value_counts()
```

```python
#filtering 
#startswith/endswith
df[df['Firstname'].str.endswith('A')]
#isdigit/isalpha
df[df['Firstname'].str.isalpha()]
```

```python
#applying regex
#contains 
#search john -> both case
df[df['Firstname'].str.contains('john',case=False)]
#find lastname starts first and end char vowel
df[df['Firstname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```
