```python
genres = movies.groupby('Genre')
genres.apply(min)
```

```python
def foo (group):
    print(group)
    return group

genres.apply(foo)
```

```python
def foo (group):    
    return group['Series_Title'].str.startswith("A").sum()    

genres.apply(foo)
```

```python
def ranking_movie(group):
    group['IMDB_Rating'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(ranking_movie)
```

```python
def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

genres.apply(normal)


#apply custome logic apply
```
